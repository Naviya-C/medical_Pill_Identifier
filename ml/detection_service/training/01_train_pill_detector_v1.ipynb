```python
import torch
from ultralytics import YOLO
```

```python
model = YOLO("yolov10m.pt")
```

```python
deviceT = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {deviceT} to train the model")
```

```python
model.train(
    data = 'data.yaml',
    epochs = 100,
    imgsz = 640,
    batch = 8,
    device = deviceT,
    patience = 20,
    mosaic=1.0,
    name = "pill_detection_1_v10m" 
)
```
